```python
import pandas as pd
import numpy as np
```

## 获取movie_id, title, type, 并去掉没有海报的电影

```python
genres = ['unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime',
          'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
          'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
i_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url',
         'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime',
          'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
          'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('../../../ml-100k/u.item', sep='|', names=i_cols, encoding='latin-1')
no_posters = np.loadtxt('./no_use.txt')
items = items.drop(columns=['release_date', 'video_release_date', 'imdb_url'])

items = items[~items['movie_id'].isin(no_posters)] # 删除没有海报的记录
items
```

## 生成训练时用到的数据

```python
with open('./data.txt', 'w') as f:  
    for index, row in items.iterrows():
        for t in range(19):
            if(row[t+2]==1):
                print(str(row[0])+'\t'+str(t)+'\t'+str(row[2:21].values)[1: -1].replace(' ', '')+'\t'+row[1])
                f.write(str(row[0])+'\t'+str(t)+'\t'+str(row[2:21].values)[1: -1].replace(' ', '')+'\t'+row[1]+'\n')
```
